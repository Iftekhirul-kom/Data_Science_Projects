```
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```
df = pd.read_csv("tic-tac-toe.csv")
df.head()
```

```
le = LabelEncoder()
df["Class"] = le.fit_transform(df["Class"])
df
```

```
#Features and label
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
y.value_counts()
```

```
def FIT(x,y):
  total = y.count()
  values = y.value_counts()

  total_0 = values[0]
  total_1 = values[1]

  FIT.P_0 = total_0 / total
  FIT.P_1 = total_1 / total

  global prob_val_arr
  prob_val_arr = []

  df_temp = pd.concat([x, y],axis=1)


  for col in x:
      uni_feat = x[col].unique()

      count_value = df_temp.groupby(col)['Class'].value_counts()
      prob_value = df_temp.groupby(col)['Class'].value_counts() / df_temp.groupby(col)['Class'].count()

      i = 0
      j = 1
      prob_val_dict = {}

      for it in uni_feat:
        prob_val_dict[it] = [prob_value[i],prob_value[j]]
        i += 2
        j += 2
      prob_val_arr.append(prob_val_dict)

      #print(prob_value)

  #return prob_val_arr
```

```
def Predict(x):
  predicted_lst = []

  for row in range(0,len(x.index)):
    idx = 0
    temp_1 = 0
    temp_0 = 0
    feat_1 = 1
    feat_0 = 1
    for each in x.iloc[row]:
      temp_1 = prob_val_arr[idx][each][0]
      feat_1 = feat_1 * temp_1

      temp_0 = prob_val_arr[idx][each][1]
      feat_0 = feat_0 * temp_0
      idx += 1

    feat_1 = feat_1 * FIT.P_1
    feat_0 = feat_0 * FIT.P_0

    if feat_1 <= feat_0:
      predicted_lst.append(feat_0)
    else:
      predicted_lst.append(feat_1)

  return predicted_lst
```

```
train_x, test_x, train_y, test_y  = train_test_split(x, y, test_size = 0.1)
train_x1, test_x1, train_y1, test_y1  = train_test_split(x, y, test_size = 0.1, stratify = y)
```

```
fit = FIT(train_x,train_y)
predict = Predict(test_x)

fit_1 = FIT(train_x1,train_y1)
predict_1 = Predict(test_x1)
```

```
acc = accuracy_score(test_y, predict)
acc_1 = accuracy_score(test_y1, predict_1)
```

```
precision = precision_score(test_y, predict)
precision_1 = precision_score(test_y1, predict_1)
```

```
recall = recall_score(test_y, predict)
recall_1 = recall_score(test_y1, predict_1)
```

```
f1 = f1_score(test_y, predict)
f1_1 = f1_score(test_y1, predict_1)
```
