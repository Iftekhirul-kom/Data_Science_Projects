```
from numpy import genfromtxt

data_path =  '/content/gdrive/My Drive/AI lab/KNN/iris.csv'
my_data = genfromtxt(data_path, delimiter=',')

data = my_data.tolist()
```

```
import random
from random import uniform
Train_set=[]
Val_set=[]
Test_set=[]

random.shuffle(data)
for S in data:
  R = uniform(0,1)
  if R>=0 and R<=0.7:
    Train_set.append(S)
  elif R>0.7 and R<=0.85:
    Val_set.append(S)
  else:
    Test_set.append(S)

```

```
import math
def euclideanDist(set1, set2):
    dist = 0
    for x in range(len(set1)-1):
        dist += pow((set1[x] - set2[x]), 2)
    return math.sqrt(dist)
```

```
#Classification
K = 20
L = []
correct = 0
total_Val = len(Val_set)
Val_acc = 0

for V in Val_set:
  for T in Train_set:
    dist = euclideanDist(V,T)
    L.append((T,dist))
  L.sort()
  K_samples = []
  for x in range(K):
        K_samples.append(L[x][0])

  majority = majority_class(K_samples)

  if (majority[-1] == V[-1]):
    correct += 1

print(correct)
Val_acc = (correct)/(total_Val) * 100
print(Val_acc)

```

```
def majority_class(num_list):
        idx, ctr = 0, 1

        for i in range(1, len(num_list)):
            if num_list[idx] == num_list[i]:
                ctr += 1
            else:
                ctr -= 1
                if ctr == 0:
                    idx = i
                    ctr = 1

        return num_list[idx]
```
