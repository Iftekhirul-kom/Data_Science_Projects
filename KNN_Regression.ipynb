```
from numpy import genfromtxt

data_path =  '/content/gdrive/My Drive/AI lab/KNN/diabetes.csv'
my_data = genfromtxt(data_path, delimiter=',')

data = my_data.tolist()
```

```
import random
from random import uniform
Train_set=[]
Val_set=[]
Test_set=[]

random.shuffle(data)
for S in data:
  R = uniform(0,1)
  if R>=0 and R<=0.7:
    Train_set.append(S)
  elif R>0.7 and R<=0.85:
    Val_set.append(S)
  else:
    Test_set.append(S)
```

```
import math
def euclideanDist(set1, set2):
    dist = 0
    for x in range(len(set1)-1):
        dist += pow((set1[x] - set2[x]), 2)
    return math.sqrt(dist)
```

```
def Average(l):
    return sum(l) / len(l)
```

```
import numpy as np
K = 5
L = []
error = 0

average = []
for V in Val_set:
  for T in Train_set:
    dist = euclideanDist(V,T)
    L.append((T,dist))
  L.sort()
  for x in range(K):
        K_samples.append(L[x][0])
  average = Average(K_samples))
  error = error + pow((V[-1]-average[-1]),2)

mean_squared_error = error/len(Val_set)






```
